```python

x = abs(int(input()))

res = 0

for i in range(x+1): 
	res += i*i
	
print(res)
```

```python
n = int(input())
m = int(input())

for i in range(1,m+1):
    tmp = n * i
    if tmp < m:
        print(tmp)
    else:
        break
    



```
