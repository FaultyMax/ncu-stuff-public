```python
import random

macierz = [[random.randint(1, 20) for _ in range(4)] for _ in range(4)]

for rzad in macierz:
    print(rzad)

def czy_istnieje(macierz, szukana):
    for rzad in macierz:
        for element in rzad:
            if element == szukana:
                print(f"{szukana} znajduje się w macierzy!")
                return True  
    print(f"{szukana} NIE znajduje się w macierzy.")
    return False

czy_istnieje(macierz, 7)

def ile_wystapien(macierz, szukana):
    licznik = 0
    for rzad in macierz:
        for element in rzad:
            if element == szukana:
                licznik += 1  
    print(f"{szukana} występuje {licznik} raz/razy.")
    return licznik

ile_wystapien(macierz, 3)
```
