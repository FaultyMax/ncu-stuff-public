```python
class Przedmiot():
    def __init__(self,nazwa,waga,wartosc):
        self.nazwa = nazwa
        self.waga = waga
        self.wartosc = wartosc

class MagicznyMiecz(Przedmiot):
    def __init__(self,nazwa,waga,wartosc,obrazenia_od_ognia):
        nazwa = "Płonący " + nazwa
        super().__init__(nazwa,waga,wartosc)
        self.obrazenia_od_ognia = obrazenia_od_ognia

    def __repr__(self):
        return f"{self.nazwa} {self.wartosc}g"


t = MagicznyMiecz("Miecz",10,250,3)
repr(t)
```

```python
class Figura:
    widocznosc = False

    def __init__(self, kolor):
        self.moj_kolor = kolor

    @classmethod
    def can_see(cls):
        if cls.widocznosc:
            return "Figury są widoczne."
        return "Figury są niewidoczne."

Figura.can_see()
Figura.widocznosc = True
Figura.can_see()
```

```python
class Termometr():

    def __init__(self):
        self._temperatura_celsius = None
    
    @property
    def temperatura_celsius(self):
        return self._temperatura_celsius

    @temperatura_celsius.setter
    def temperatura_celsius(self, val):
        if val < -273.15:
            raise RuntimeError("Ponizej zera absolutnego!")
        self._temperatura_celsius = val

t = Termometr()
print(t._temperatura_celsius)
t.temperatura_celsius = 20
print(t._temperatura_celsius)
t.temperatura_celsius = -5550
```

```python
class Produkt:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self._cena = None

    @property
    def cena(self):
        return self._cena

    @cena.setter
    def cena(self, val):
        if val < 0:
            raise RuntimeError("Ujemna!")
        self._cena = val

p = Produkt("pudelko")
print(p.cena)
p.cena = 20
print(p.cena)
p.cena = -2000
```

```python
class Kalkulator():

    licznik_instancji = 0
    
    def __init__(self):
        Kalkulator.licznik_instancji += 1

    @staticmethod
    def dziel_bezpiecznie(a,b):
        if b != 0:
            return a/b
        else:
            raise RuntimeError("dzielenie przez 0")

    @classmethod
    def daj_licznik(cls):
        return cls.licznik_instancji

for i in range(5):
    Kalkulator()


print(Kalkulator.daj_licznik())
#Kalkulator.dziel_bezpiecznie(10,0)

moj_kalkulator = Kalkulator()

#a,b  = input(),input()

a = "wersja_softu"
b = "2.0"

setattr(moj_kalkulator, a, b)
print(getattr(moj_kalkulator, a))

Kalkulator.__dict__
moj_kalkulator.__dict__
```

```python
from dataclasses import dataclass, field

@dataclass
class PomiarPodstawowy:
    stacja: str
    temperatura: float
    wilgotnosc: float = field(default="Brak danych")
    order=True
    frozen=True

    def czy_upal(self) -> bool:
        if self.temperatura > 30.0:
            return True
        return False

w = PomiarPodstawowy("Toruń",3.14)
w
```

```python
from pydantic import BaseModel, StrictInt, StrictStr, StrictFloat

class PomiarZwalidowany(pydantic.BaseModel):
    
    stacja: StrictStr
    temperatura: StrictFloat


a = PomiarZwalidowany(stacja="Toruń",temperatura="25.5")
a
```
