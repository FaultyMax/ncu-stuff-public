```python
from functools import wraps

def powtorz(k=1):                  # 1. Przyjmuje argumenty dekoratora
  def wlasciwy_dekorator(func):    # 2. Przyjmuje funkcję do udekorowania
     @wraps(func)                  # Zawsze warto zachować metadane!
     def wrapper(*args, **kwargs): # 3. Zastępuje oryginalną funkcję
        wynik = None
        for i in range(k):
          print(f"Wykonanie {i+1} z {k}")
          wynik = func(*args, **kwargs)
        return wynik               # Zwracamy wynik ostatniego wywołania
     return wrapper                # Zwraca wrapper
  return wlasciwy_dekorator        # Zwraca dekorator

@powtorz(k=3)
def powitaj(imie):
    print(f"Cześć, {imie}!")

powitaj("Toruń")
```

```python
import time

def mierz_czas(func):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        wynik = func(*args, **kwargs) # Wywołanie oryginalnej funkcji
        koniec = time.perf_counter()
        print(f"Funkcja {func.__name__} działała {koniec - start:.6f} s")
        return wynik
    return wrapper

@mierz_czas
def wolna_funkcja(n):
    return sum(i**2 for i in range(n))

wolna_funkcja(1_000_000)
```

```python
def logger(func):
    @wraps(func)
    def wrapper(*args,**kwargs):
        print("Rozpoczynam wywołanie...")
        wynik = func(*args, **kwargs)
        print("Zakończono wywołanie")
        return wynik
    return wrapper

@logger
def fibonacci(n):
    sequence = [0, 1]
    for i in range(2, n):
        next_val = sequence[-1] + sequence[-2]
        sequence.append(next_val)
    #print(sequence)
    return sequence[:n]

fibonacci(10)
```

```python

def wymus_wielkie_litery(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        wynik = func(*args, **kwargs)
        if isinstance(wynik,str):
            wynik = wynik.upper()
        return wynik
    return wrapper

@wymus_wielkie_litery
def funkcja(napis):
    return "do tego dodam " + napis

funkcja("x")
```

```python

def sprawdz_typ_str(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        #print(args)
        for arg in args:
            if not isinstance(arg,str):
                #print(f" nie string {arg}")
                raise TypeError("argumenty musza byc tylko ciagami znakow") 
        wynik = func(*args, **kwargs)
        return wynik
    return wrapper

@sprawdz_typ_str
def argumenty(a, b, c, *args, **kwargs):
    pass

argumenty("sd","t","aa", n = 4)
```

```python
import time

def opoznienie(sekundy=1):
    def dek_opoznienie(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            time.sleep(sekundy)
            wynik = func(*args,**kwargs)
            return wynik
        return wrapper
    return dek_opoznienie

@opoznienie(4)
def wolna_funkcja(n):
    return sum(i**2 for i in range(n))

wolna_funkcja(4)
```
