```python
def oblicz_lokate(kwota_poczatkowa, oprocentowanie, lata):
    """
    Oblicz lokate przy danym oprocentowaniu i czasie.

    :param kwota_poczatkowa: Kwota.
    :type kwota_poczatkowa: float
    :param oprocentowanie: Oprocentowanie.
    :type oprocentowanie: float
    :param lata: Lata.
    :type lata: int
    :raises ValueError: Jeśli `lata` jest ujemne.
    :return: wartość końcowa lokaty przy procentowaniu składanym
    :rtype: float

    """
    if lata < 0:
        raise ValueError("lata nie może być ujemne")
    return kwota_poczatkowa * (1 + oprocentowanie) ** lata
```

```python
class Odliczanie:

    def __init__(self, start):
        self.start = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.start < 0:
            raise StopIteration

        tmp = self.start
        self.start -= 1
        return str(tmp)


for i in Odliczanie(3):
    print(i)
```

```python
class Parzyste:

    def __init__(self, limit):
        self.limit = limit
        self.start = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.start > self.limit:
            raise StopIteration

        tmp = self.start
        self.start += 2
        return str(tmp)


for i in Parzyste(10):
    print(i)
```

```python
class Kwadraty:

    def __init__(self, n):
        self.n = n

    def __len__(self):
        return self.n

    def __getitem__(self, idx):
        if idx >= self.n:
            raise IndexError("index bigger than n")
        return idx * idx


k = Kwadraty(5)

for i in k:
    print(i)
```

```python
from itertools import combinations

my_set = [1, 2, 3, 4, 5]

count = 0

for i, j in combinations(my_set, 2):
    count += 1
    print(f"{i}{j}", end=", ")
print(f"\n{count}")
```

```python
from itertools import permutations

count = 0

for i, j, k in permutations("ABCD", 3):
    count += 1
    print(f"{i}{j}{k}", end=", ")
print(f"\n{count}")
```

```python
from itertools import zip_longest

list1 = [1, 2, 3]
list2 = ["A", "B"]

zipped_custom = list(zip_longest(list1, list2, fillvalue="Nieznane"))
print(zipped_custom)
```
