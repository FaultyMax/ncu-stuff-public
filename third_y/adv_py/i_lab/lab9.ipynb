```python
def sum_to_n(n):
    s = 0
    i = 0
    while (i := i + 1) <= n:
        s += i
    return s


sum_to_n(10)
```

```python
match input().split():
    case [a, "+", b]:
        print(int(a) + int(b))
    case "-":
        print(a - b)
    case "*":
        print(a * b)
    case "/":
        if b == 0:
            print("Dzielenie przez zero")
        else:
            print(a // b)
    case _:
        print("Unknown status")
```

```python
triangles = [
    (5, 5, 5),  # Równoboczny
    (5, 5, 8),  # Równoramienny
    (3, 4, 5),  # Różnoboczny
    (1, 2, 10),  # Nieprawidłowy (nie spełnia warunku trójkąta)
    (0, 0, 0),  # Nieprawidłowy
]


def check(data):
    match data:
        case (a, b, c) if a + b < c:
            print("Nie spełnia warunku trójkąta.")
        case (a, b, c) if a == b == c:
            print("Trójkąt równoboczny.")
        case (a, b, c) if a == b:
            print("Trójkąt równoramienny.")
        case _:
            print("Nieprawidłowy")


check(triangles[1])
```
