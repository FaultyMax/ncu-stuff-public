```python
import pickle

slownik = {"k1": 12, "k2": "abc", "k3": range(10)}

for i in range(6):
    print(f"wielkosc pliku dla protokolu {i} : {len(pickle.dumps(slownik,i))}")

with open("dane.pkl", "wb") as f:
    pickle.dump(slownik, f, 1)


with open("dane.pkl", "rb") as f:
    wczytany_slownik = pickle.load(f)


print(wczytany_slownik)
```

```python
class uzytkownik:

    def __init__(self, imie, nazwisko, wiek):
        self.imie = imie
        self.nazwisko = nazwisko
        self.wiek = wiek
        self.czy_pelnoletni = True if self.wiek >= 18 else False

    def __getstate__(self):
        stan = self.__dict__.copy()
        del stan["czy_pelnoletni"]
        return stan

    def __setstate__(self, d):
        self.__dict__ = d
        d["czy_pelnoletni"] = True if self.wiek >= 18 else False


example = uzytkownik("Jan", "Kowalski", 25)

a = pickle.dumps(example)
#  print(a)

b = pickle.loads(a)
print(f"{b.imie} {b.nazwisko} {b.wiek} {b.czy_pelnoletni}")
```

```python
import xml.etree.ElementTree as ET

xdata = """<?xml version="1.0"?>
<data source="Internet">
  <country name="Liechtenstein">
      <year>2008</year>
      <gdppc>141100</gdppc>
      <neighbor name="Austria" direction="E"/>
      <neighbor name="Switzerland" direction="W"/>
  </country>
  <country name="Singapore">
      <year>2011</year>
      <gdppc>59900</gdppc>
      <neighbor name="Malaysia" direction="N"/>
  </country>
</data>"""


tree = ET.fromstring(xdata)

for child in tree:
    print(child.tag, child.attrib)
    for inner_kid in child:
        print(inner_kid.tag, inner_kid.attrib)

print(tree[0][1].text)

for country in tree.findall("country"):
    print(country.get("name"))
```

```python
def stworz_html(tytul, tresc):
    root = ET.Element("html")

    head = ET.SubElement(root, "head")
    title = ET.SubElement(head, "title")
    title.text = tytul

    body = ET.SubElement(root, "body")
    body.set("bgcolor", "#ffffff")
    if type(tresc) == list:
        for text in tresc:
            paragraph = ET.SubElement(body, "p")
            paragraph.text = text
    else:
        body.text = tresc

    print(ET.tostring(root, encoding="unicode"))


stworz_html("strona", ["a", "b", "c"])
```

```python
def przelicz_proporcje(waga_bazowa: int, mnoznik: float) -> str:
    # Funkcja powinna zwrócić nową wagę w gramach
    nowa_waga = waga_bazowa * mnoznik
    return "Nowa waga: " + str(nowa_waga) + "g"


# Testowanie funkcji
wynik1 = przelicz_proporcje(200, 1.5)
wynik2 = przelicz_proporcje("300", 2)

print(wynik1)
print(wynik2)
```
